```python
import numpy as np
```

```python
# 构造特征
a = np.random.randn(100)
print(np.var(a))
```

```python
b = np.random.randn(100)*0.1
b = np.random.normal(5,0.1,100)
print(np.var(b))
```

```python
# 构造特征向量(输入数据X)
X = np.vstack((a,b)).T
print(X)
print(X.shape)
```

```python
# 低方差过滤
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(0.01)
X_filterd = vt.fit_transform(X)
```
